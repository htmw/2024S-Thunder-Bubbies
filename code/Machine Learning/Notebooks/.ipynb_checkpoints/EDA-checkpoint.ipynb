```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from glob import glob
import seaborn as sns
import matplotlib.image as mpimg
from PIL import Image
```

```python
base_dir = '../Dataset'
```

```python
images_dir = os.path.join(base_dir, 'images')
```

```python
imageid_path_dict = {os.path.splitext(os.path.basename(x))[0]: x for x in glob(os.path.join(images_dir, '*.jpg'))}
```

```python
lesion_type_dict = {
    'NV': 'Melanocytic nevi',
    'MEL': 'dermatofibroma',
    'BKL': 'Benign keratosis-like lesions ',
    'BCC': 'Basal cell carcinoma',
    'AKIEC': 'Actinic keratoses',
    'VASC': 'Vascular lesions',
    'DF': 'Dermatofibroma'
}
```

```python
df = pd.read_csv("../Dataset/GroundTruth.csv")
df.head()
```

```python
df['image_path'] = ''

for index, row in df.iterrows():
    image_id = row['image']

    if image_id in imageid_path_dict:
        df.at[index, 'image_path'] = imageid_path_dict[image_id]
updated_csv_path = os.path.join(base_dir, 'UpdatedGroundTruth.csv')
df.to_csv(updated_csv_path, index=False)
```

```python
sns.set_style("whitegrid")

label_columns = ['MEL', 'NV', 'BCC', 'AKIEC', 'BKL', 'DF', 'VASC'] 
for label in label_columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=label, data=df)
    plt.title(f'Distribution of {label}')
    plt.show()
```

```python
def get_disease(row):
    diseases = ['MEL', 'NV', 'BCC', 'AKIEC', 'BKL', 'DF', 'VASC']
    
    for disease in diseases:
        if row[disease] == 1:
            return disease
    
    return 'None or Healthy'
```

```python
df['Disease'] = df.apply(get_disease, axis=1)
```

```python
diseases = ['NV', 'MEL', 'BKL', 'BCC', 'AKIEC', 'VASC', 'DF']

images_per_disease = 6

plt.figure(figsize=(15, 10))

subplot_num = 1

for disease in diseases:
    disease_df = df[df['Disease'] == disease].head(images_per_disease)

    for _, row in disease_df.iterrows():
        plt.subplot(len(diseases), images_per_disease, subplot_num)
        
        img = mpimg.imread(row['image_path'])
        plt.imshow(img)
        
        plt.axis('off')
        
        if subplot_num % images_per_disease == 1:
            plt.title(f"{disease}\n{row['image']}")
        else:
            plt.title(row['image'])
        
        subplot_num += 1

plt.tight_layout()

plt.show()
```

```python
import os

image_dimensions = []

for _, row in df.iterrows():
    with Image.open(row['image_path']) as img:
        image_dimensions.append(img.size)  
```

```python
dimensions_df = pd.DataFrame(image_dimensions, columns=['Width', 'Height'])

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(data=dimensions_df, x='Width', bins=30, kde=True)
plt.title('Distribution of Image Widths')

plt.subplot(1, 2, 2)
sns.histplot(data=dimensions_df, x='Height', bins=30, kde=True)
plt.title('Distribution of Image Heights')

plt.tight_layout()
plt.show()
```

```python
def get_average_color(image_path):
    with Image.open(image_path) as img:
        np_img = np.array(img)
        return np.mean(np_img, axis=(0, 1))
```

```python
sample_df = df.sample(n=100, random_state=42)
sample_df['average_color'] = sample_df['image_path'].apply(get_average_color)

sample_df[['avg_red', 'avg_green', 'avg_blue']] = pd.DataFrame(sample_df['average_color'].tolist(), index=sample_df.index)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=sample_df, x='avg_red', y='avg_green', hue='Disease', style='Disease', palette='bright', s=100, alpha=0.6)
plt.title('Average Image Color by Disease')
plt.xlabel('Average Red Intensity')
plt.ylabel('Average Green Intensity')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
